#### Vectorized version runs 300 times faster than non-vectorized version. Whenever possible, avoid explicit for-loops

```python
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()

print(c)
print("Vectorized version:" + str(1000*(toc-tic)) +"ms")
```

```python
c = 0
tic = time.time()
for i in range(1000000):
    c += a[i]*b[i]
toc = time.time()

print(c)
print("For loop:" + str(1000*(toc-tic)) + "ms")
```

#### Broadcasting Example

```python
import numpy as np

A = np.array([[56.0,0.0,4.4,68.0],
              [1.2,104.0,52.0,8.0],
              [1.8,135.0,99.0,0.9]])

print(A)
```

```python
cal = A.sum(axis=0) # axis=0, sums vertically
print(cal)
```

```python
percentage = 100*A/cal
print(percentage)
```
